```python
import json 
import pandas as pd 
import ast
```

```python
with open("/Users/marisa/clausal-coordinate-ellipsis/german-common-crawl/de_head_0000_2015-48.jsonl", "r") as file:
    lines = list(file)
```

```python
dict_data = []
for line in lines:
    dict_data.append(ast.literal_eval(line))
```

```python
result = [] # enthält ordentliches json
for d in dict_data:
    result.append(json.dumps(d))
```

```python
dict_list = [] # enthält dictionaries aus json
for r in result:
    dict_list.append(json.loads(r))
```

```python
df = pd.DataFrame(dict_list) # dataframe tabelle
```
